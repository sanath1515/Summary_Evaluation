```python
def countWords(description):
    return len(description.split())
```

```python
def distinctWords(description):
    d={}
    for i in description.split():
        if(i in d):
            d[i] = d[i]+1
        else:
            d[i] = 1
            
    c = 0
    for i in d:
        if d[i] == 1:
            c+=1
    
    return c
distinctWords("Hello the the")
```

```python
def commonWords(a,b):
    return len(set(a.split()).intersection(set(b.split())))
commonWords("The India","The wgl")
```

```python
def avg_word_length(description):
    count = 0
    for i in description.split():
        count += len(i)
    return count/len(description.split())

avg_word_length("abc de fghi")
```

```python
def countSentences(desc):
    return max(desc.count('.'),1)

countSentences("the. abc. lmn. raf. jghf")
```

```python
def countSyllables(word):
    word = word.lower()
    count = 0
    vowels = "aeiouy"
    if word[0] in vowels:
        count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            count += 1
    if word.endswith("e"):
        count -= 1
    if count == 0:
        count += 1
    return count

print(countSyllables('meat'))
```

```python
# by Flesch Reading Ease
def calculate_readability(desc):
    return 206.835 - 1.015 * (countWords(desc)/countSentences(desc)) - 84.6 * (countSyllables(desc)/countWords(desc))
calculate_readability("The third wave was an experimentto see how people reacted to a new one leader government. It gained popularity as people wanted to try new things. The students follow anything that is said and start turning on eachother to gain higher power. They had to stop the experement as too many people got to radical with it blindly following there leader")
```

```python
def avg_sentence_length(desc):
    count = 0
    for i in desc.split('.'):
        count += len(i)
#         print(i,len(i))
        
    return count/len(desc.split('.'))

avg_sentence_length("the hello.abc bfbidsfg.lmn fherh.raf dgdh.jghf dbdnorh")
```

```python
from string import punctuation
def punctuation_denstiy(desc):
    count = 0
    for i in desc:
        if i in punctuation:
            count = count + 1
    return count/len(desc.split("."))

sentence = "I love cricket, football , hockry !!.the, hi."
punctuation_denstiy(sentence)
```

```python
# sentences must be array of sentence
def calculate_conjunction_density(sentences):
    # Tokenize the sentences into individual words.
    tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
    
    # Tag each word with its part of speech.
    tagged_sentences = [nltk.pos_tag(sentence) for sentence in tokenized_sentences]
    
    # Identify all the conjunctions in the list of tagged words.
    conjunctions = [word for sentence_tags in tagged_sentences for word, tag in sentence_tags if tag in ["CC", "IN"]]
    
    # Calculate the percentage of conjunctions in the list of words.
    conjunction_percentage = len(conjunctions) / sum(len(sentence) for sentence in tokenized_sentences)
    return conjunction_percentage

sentences = ["I like to eat apples and oranges.", "I like to swim and play basketball."]
calculate_conjunction_density(sentences)
```

```python
def good_bad_words(good_essays,bad_essays):
    vectorizer = TfidfVectorizer()
    tfidf_good = vectorizer.fit_transform(good_essays)
    tfidf_bad = vectorizer.transform(bad_essays)
    feature_names = vectorizer.get_feature_names_out()
    avg_tfidf_good = tfidf_good.mean(axis=0)
    avg_tfidf_bad = tfidf_bad.mean(axis=0)
    df = pd.DataFrame({'Word': feature_names, 'Avg_TF-IDF_Good': avg_tfidf_good.tolist()[0], 'Avg_TF-IDF_Bad': avg_tfidf_bad.tolist()[0]})
    df['TF-IDF_Difference'] = df['Avg_TF-IDF_Good'] - df['Avg_TF-IDF_Bad']
    df = df.sort_values(by='TF-IDF_Difference', ascending=False)
    print("Top Over-Represented Words in Good Essays:")
    print(df.head(10))
    print("Top Under-Represented Words in Bad Essays (Conversely):")
    print(df.tail(10))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#features []
#df data frame
#targets []
#test_size 0.2
#random_state 42

def buildModelAndReturnMSE(df,features,targets,test_size,random_state):
    # Define the original five features and target variables
    features = features
    target_content = targets[0]
    target_wording = targets[1]
    
    # Split the data into training (80%) and testing (20%) sets
    X_train, X_test, y_train_content, y_test_content, y_train_wording, y_test_wording = train_test_split(
    df[features], df[target_content], df[target_wording], test_size=test_size, random_state=random_state)
    
    # Initialize linear regression models
    content_model = LinearRegression()
    wording_model = LinearRegression()
    
    # Train the models on the training data
    content_model.fit(X_train, y_train_content)
    wording_model.fit(X_train, y_train_wording)
    
    # Make predictions on the testing data
    y_pred_content = content_model.predict(X_test)
    y_pred_wording = wording_model.predict(X_test)
    
    # Calculate mean squared error for content and wording models
    mse_content = mean_squared_error(y_test_content, y_pred_content)
    mse_wording = mean_squared_error(y_test_wording, y_pred_wording)
    
    # Report mean squared error for each model
    print("Mean Squared Error (Content Model):", mse_content)
    print("Mean Squared Error (Wording Model):", mse_wording)
    
    
```

```python
features = ["text_word_count", "prompt_text_word_count", "common_text_and_prompt_text", "common_text_and_prompt_question", "common_text_and_prompt_title"]
targets = ["content", "wording"]
test_size = 0.2
random_state = 99
buildModelAndReturnMSE(final_features_df,features,targets,test_size,random_state)
```
